```python
import numpy as np
import pandas as pd
import re
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
# Read/Load the dataset files
# TODO
train_data = pd.read_csv('train_dataset.csv')
test_data = pd.read_csv('test_dataset.csv')
train_data.head()
```

```python
# Preprocess the dataset
# TODO
train_data.drop(['Distribution', 'Habits', 'Mating_Habits', 'Population', 'Order', 'Suborder', 'Family', 'Genus', 'Species'], axis=1, inplace=True)
test_data.drop(['Distribution', 'Habits', 'Mating_Habits', 'Population'], axis=1, inplace=True)

train_data['Kingdom'] = train_data['Kingdom'].astype('category').cat.codes
train_data['Phylum'] = train_data['Phylum'].astype('category').cat.codes
train_data['Subphylum'] = train_data['Subphylum'].astype('category').cat.codes
train_data['Attributes'] = train_data['Attributes'].astype('category').cat.codes
train_data['Diet'] = train_data['Diet'].astype('category').cat.codes
test_data['Kingdom'] = test_data['Kingdom'].astype('category').cat.codes
test_data['Phylum'] = test_data['Phylum'].astype('category').cat.codes
test_data['Subphylum'] = test_data['Subphylum'].astype('category').cat.codes
test_data['Attributes'] = test_data['Attributes'].astype('category').cat.codes
test_data['Diet'] = test_data['Diet'].astype('category').cat.codes

for column in ['Weight', 'Height', 'Length', 'Life span', 'Top speed']:
    for element in train_data[column]:
        if type(element) == str:
            numbers = list(map(int, re.findall(r'\d+', element)))
            if len(numbers) == 0:
                train_data[column].replace(element, np.nan, inplace=True)
                print(column)
            train_data[column].replace(element, sum(numbers)/len(numbers), inplace=True)
            
    
    for element in test_data[column]:
        if type(element) == str:
            numbers = list(map(int, re.findall(r'\d+', element)))
            test_data[column].replace(element, sum(numbers)/len(numbers), inplace=True)
            
    train_data[column] = train_data[column].astype(float)
    test_data[column] = test_data[column].astype(float)
            
_class = train_data['Class']
train_data.drop('Class', axis=1, inplace=True)

def find_median(list_):
    _list = []
    for elem in list_:
        if elem == elem:
            _list.append(elem)
    return sorted(_list)[len(_list)//2]

def find_mean(list_):
    _list = []
    for elem in list_:
        if elem == elem:
            _list.append(elem)
    return sum(_list) / len(_list)

for column in train_data.columns:
    if train_data[column].isna().sum() == len(train_data[column]) or test_data[column].isna().sum() == len(test_data[column]):
        train_data.drop(column, axis=1, inplace=True)
        test_data.drop(column, axis=1, inplace=True)
        
def parse_population(value):
    if pd.isna(value) or value.strip().lower() == 'unknown':
        return np.nan
    
    value = value.lower()
    
    if 'below' in value or 'under' in value or '<' in value:
        numbers = re.findall(r'\d+\.?\d*', value.replace(',', ''))
        if numbers:
            return float(numbers[0])
        else:
            return np.nan
    elif 'over' in value or 'above' in value or '>' in value:
        numbers = re.findall(r'\d+\.?\d*', value.replace(',', ''))
        if numbers:
            return float(numbers[0])
        else:
            return np.nan
    
    if '-' in value:
        parts = value.split('-')
        if len(parts) == 2:
            num1 = parse_single_value(parts[0])
            num2 = parse_single_value(parts[1])
            if not np.isnan(num1) and not np.isnan(num2):
                return (num1 + num2) / 2 
            else:
                return np.nan
    
    return parse_single_value(value)

def parse_single_value(value):
    value = value.replace(',', '').replace(' ', '')
    if 'thou' in value or 'k' in value:
        multiplier = 1e3
        value = value.replace('thou', '').replace('k', '')
    elif 'mln' in value or 'm' in value:
        multiplier = 1e6
        value = value.replace('mln', '').replace('m', '')
    elif 'b' in value:
        multiplier = 1e9
        value = value.replace('b', '')
    else:
        multiplier = 1
    
    numbers = re.findall(r'\d+\.?\d*', value)
    if numbers:
        return float(numbers[0]) * multiplier
    else:
        return np.nan
    
train_data['Population size'] = train_data['Population size'].apply(parse_population)
test_data['Population size'] = test_data['Population size'].apply(parse_population)

for column in train_data.columns:
    if column == 'Population size':
        _mean = find_mean(train_data[column])
        for i in range(len(train_data[column])):
            if train_data[column][i] != train_data[column][i]:
                train_data[column][i] = _mean
        _mean = find_mean(test_data[column])
        for i in range(len(test_data[column])):
            if test_data[column][i] != test_data[column][i]:
                test_data[column][i] = _mean
    else:
        if train_data[column].isna().sum() > 0:
            median = find_median(train_data[column])
            for i in range(len(train_data[column])):
                if train_data[column][i] != train_data[column][i]:
                    train_data[column][i] = median
        
        if test_data[column].isna().sum() > 0:
            median = find_median(test_data[column])
            for i in range(len(test_data[column])):
                if test_data[column][i] != test_data[column][i]:
                    test_data[column][i] = median

train_data['Population size'] = (train_data['Population size'] - train_data['Population size'].mean()) / train_data['Population size'].std()


train_data.head(100)
```

```python
# Design and train your model
# TODO
best_model, best_acc = None, None
X_train, X_val, y_train, y_val = train_test_split(train_data, _class, test_size=0.2, random_state=42)
# for n_estimators in range(5, 100, 5):
#     for max_depth in range(5, 20, 5):
#         model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)
#         model.fit(X_train, y_train)
#         y_pred = model.predict(X_val)
#         acc = f1_score(y_val, y_pred, average='weighted')
#         if best_acc is None or acc > best_acc:
#             best_acc = acc
#             best_model = model
            
# AdaBoost model
from sklearn.ensemble import AdaBoostClassifier
for n_estimators in range(1000, 1001, 5):
    model = AdaBoostClassifier(n_estimators=n_estimators, learning_rate=.1)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    acc = f1_score(y_val, y_pred, average='weighted')
    if best_acc is None or acc > best_acc:
        best_acc = acc
        best_model = model
```
